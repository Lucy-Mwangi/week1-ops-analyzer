### Objective
This notebook automates efficiency tracking across 3 operational sites.
Instead of manually calculating performance figures in a spreadsheet,
we use Python to instantly score each site and flag any that need attention.

Think of this notebook as a digital performance monitor — it reads the
numbers, does the maths, and tells a manager exactly which sites are
performing well and which ones need urgent action.

---
## Step 1: The KPI Function
### What is this?
A **function** in Python is like a recipe — you give it ingredients
(inputs), it follows a set of steps, and gives you a result (output).

Here we write a recipe called `calculate_efficiency`.  
You give it two numbers:
- **actual_output** — what the site actually produced
- **target_output** — what the site was supposed to produce

It gives back one number: the **efficiency percentage** — how close
the site got to its goal, expressed as a percentage.

*Example: If a depot was meant to process 10,000 units but only
processed 8,750 — the efficiency is 87.5%*

# ============================================================
# FUNCTION 1: calculate_efficiency
# Purpose: Calculate how efficiently a site met its target
# ============================================================

def calculate_efficiency(actual_output, target_output):
    """
    Calculates the efficiency percentage of an operational site.
    
    Parameters:
        actual_output  : The real amount the site produced
        target_output  : The amount the site was supposed to produce
    
    Returns:
        Efficiency as a percentage (float), rounded to 2 decimal places
    """
    efficiency = (actual_output / target_output) * 100
    return round(efficiency, 2)


# --- Quick Test ---
test_efficiency = calculate_efficiency(8750, 10000)
print(f"Test Result: {test_efficiency}%")
print("✅ KPI Function is working correctly.")

---
## Step 2: The Status Classifier
### What is this?
Once we have the efficiency score, we need to translate that number
into a clear label that any manager can act on immediately.

This second recipe — `get_operational_status` — takes the efficiency
percentage and returns one of three plain-English labels:

| Efficiency Score | Status Label | What it means |
|---|---|---|
| 90% and above | ✅ Normal | Site is performing well |
| 70% to 89% | ⚠️ Warning | Site needs attention |
| Below 70% | 🔴 Critical | Immediate action required |

This removes all guesswork — the system tells you exactly where
to focus your energy.

# ============================================================
# FUNCTION 2: get_operational_status
# Purpose: Classify a site's performance based on efficiency
# ============================================================

def get_operational_status(efficiency):
    """
    Classifies operational status based on efficiency percentage.
    
    Parameters:
        efficiency : The efficiency percentage (float)
    
    Returns:
        A status string: "Critical", "Warning", or "Normal"
    """
    if efficiency < 70:
        return "🔴 Critical"
    elif efficiency < 90:
        return "⚠️  Warning"
    else:
        return "✅ Normal"


# --- Quick Test ---
print(get_operational_status(51.0))   # Expected: Critical
print(get_operational_status(87.5))   # Expected: Warning
print(get_operational_status(102.0))  # Expected: Normal
print("✅ Status Classifier is working correctly.")

## Step 3: Data Modelling — 3 Operational Sites
### What is this?
We now create a **data model** — a structured, organised way of
storing information about each site in one place.

In Python, we use something called a **dictionary** — think of it
like a digital filing cabinet. Each drawer (site) is clearly labelled
and contains exactly the same type of information:
- The site's name
- Its actual output this month
- Its target output for the month

We are modelling **3 Kenya Pipeline Company fuel depots** — Nairobi,
Mombasa, and Kisumu — tracking how many units of fuel each depot
processed against its monthly target.

# ============================================================
# DATA MODEL: 3 Operational Fuel Depot Sites
# Industry: Energy / Pipeline Operations (KPC)
# ============================================================

operational_sites = {
    "Site_A": {
        "name"          : "Nairobi Depot",
        "actual_output" : 8750,    # units processed this month
        "target_output" : 10000    # monthly target units
    },
    "Site_B": {
        "name"          : "Mombasa Depot",
        "actual_output" : 10200,
        "target_output" : 10000
    },
    "Site_C": {
        "name"          : "Kisumu Depot",
        "actual_output" : 5100,
        "target_output" : 10000
    }
}

print(f"✅ Data loaded. Monitoring {len(operational_sites)} sites.")

---
## Step 4: Generating the Operational Report
### What is this?
Now we bring everything together.

We use a **loop** — an instruction that says *"do this same process
for every site, one by one, automatically."* Instead of manually
calculating each site's score one at a time, the loop does it
instantly across all sites.

For each site, the loop will:
1. Pick up the site's numbers from our filing cabinet
2. Run them through the efficiency calculator (Step 1)
3. Pass the result to the status classifier (Step 2)
4. Print a clean, formatted summary

This is the equivalent of a manager reviewing 3 monthly reports
simultaneously — in under one second.

# ============================================================
# EXECUTION: Loop through all sites and generate report
# ============================================================

print("=" * 55)
print("    KPC OPERATIONAL PERFORMANCE REPORT")
print("    Inuka Tech Fellowship | Week 1 Analysis")
print("=" * 55)

for site_id, site in operational_sites.items():
    
    # Step 1: Calculate efficiency for this site
    efficiency = calculate_efficiency(
        site["actual_output"],
        site["target_output"]
    )
    
    # Step 2: Classify the operational status
    status = get_operational_status(efficiency)
    
    # Step 3: Print formatted report for this site
    print(f"\n📍 Site          : {site['name']}")
    print(f"   Target Output : {site['target_output']:,} units")
    print(f"   Actual Output : {site['actual_output']:,} units")
    print(f"   Efficiency    : {efficiency}%")
    print(f"   Status        : {status}")

print("\n" + "=" * 55)
print("✅ Report Complete. Review Critical sites immediately.")
print("=" * 55)
